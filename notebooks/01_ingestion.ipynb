```python
processed_dir = "data/processed/arxiv"
os.makedirs(processed_dir, exist_ok=True)

texts = {}

for pdf in pdf_paths:
    print(f"Parsing: {pdf}")
    text = parse_pdf_to_text(pdf)
    
    out_path = os.path.join(processed_dir, os.path.basename(pdf).replace(".pdf", ".txt"))
    with open(out_path, "w") as f:
        f.write(text)
    
    texts[pdf] = text
```
