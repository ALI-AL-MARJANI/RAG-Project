```python
import os
import json
from src.processing.parser import parse_pdf

RAW_DIR = "data/raw/arxiv"
OUT_DIR = "data/processed/parsed"
os.makedirs(OUT_DIR, exist_ok=True)

pdf_files = [f for f in os.listdir(RAW_DIR) if f.endswith(".pdf")]
pdf_files
```

```python
sample = pdf_files[0]
sample_path = os.path.join(RAW_DIR, sample)

text = parse_pdf(sample_path)

# Preview
text[:2000]
```

```python
def save_document(base_name, text, out_dir):
    txt_path = os.path.join(out_dir, base_name + ".txt")
    json_path = os.path.join(out_dir, base_name + ".json")

    with open(txt_path, "w") as f:
        f.write(text)

    with open(json_path, "w") as f:
        json.dump({"id": base_name, "content": text}, f, indent=2)

    return txt_path, json_path


for pdf in pdf_files:
    base = pdf.replace(".pdf", "")
    text = parse_pdf(os.path.join(RAW_DIR, pdf))

    save_document(base, text, OUT_DIR)

print("Parsing terminé et fichiers sauvegardés.")
```
